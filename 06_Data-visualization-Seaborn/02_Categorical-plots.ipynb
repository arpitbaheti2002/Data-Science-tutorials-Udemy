# Categorical Data Plots
Some categorical plots are:
+ boxplot
+ violinplot
+ stripplot
+ swarmplot
+ barplot
+ countplot

<hr>

```python
import seaborn as sns
%matplotlib inline
```

```python
data = sns.load_dataset('titanic')
data
```

<hr>

## barplots and countplots

**Barplot** is a general plot that allows you to aggregate the categorical data based off some function, by default the mean:

```python
sns.barplot(x='sex', y='fare', data=data)
```

```python
# we can change the estimator objec

import numpy as np
sns.barplot(x='sex', y='fare', data=data, estimator=np.sum)
```

**countplot** is essentially the same as barplot except the estimator is explicitly counting the number of occurrences

```python
sns.countplot(x='sex', data=data)
```

```python
sns.countplot(x='sex', data=data[data['survived']==1], hue='class')
```

<hr>

## boxplot and violinplot

A **boxplot** (or box-and-whisker plot) shows the distribution of qunatative datain a way that facilitates comparison between variables or across levels of a categorical variable. The box shows the quartiles of the datasetwhile the whiskers extend to show the rest of the distribution, except for points that are determined to be "outliers" using a method that is a function of the inter-quartile range.

```python
sns.boxplot(x='class', y='age', data=data, palette="rainbow")
```

```python
# Can do entire dataframe with orient='h'
sns.boxplot(x='age' ,data=data ,palette='rainbow',orient='h')
```

```python
sns.boxplot(x='class', y='age', data=data[data['survived']==1], hue='sex')
```

A **violin plot** show the distribution of quantitative data acress several levels of one(or more) categorical variales such that those distribution can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution. 

```python
sns.violinplot(x='sex', y='age', data=data[data['survived']==1])
```

```python
sns.violinplot(x='sex', y='age', data=data[data['survived']==1], hue='class')
```

```python
sns.violinplot(x='class', y='age', data=data[data['survived']==1], hue='sex', split=True)
```

<hr>

## stripplot and swarmplot

The stripplot draws a scatterplot where one variable is categorical. It can be well complemented with box plot or violin plot.

```python
sns.stripplot(x='class', y='age', data=data)
```

```python
sns.stripplot(x='class', y='age', data=data, hue='sex')
```

```python
sns.swarmplot(x='class', y='age', data=data, size=3.5)
```

```python
sns.swarmplot(x='class', y='age', data=data, hue='alive', size=3.5)
```

```python
sns.swarmplot(x='class', y='age', data=data, hue='alive', dodge=True, size=4)
```

<hr>

## Combining Categorical Plots

```python
sns.swarmplot(x="sex", y='age', data=data, size=3, color='black')
sns.violinplot(x="sex", y='age', data=data)
```

<hr>

## catplot
catplot is the most general form of a categorical plot. It can take in a **kind** paramenter to adjust the plot type:

```python
sns.catplot(x='sex', y='fare', data=data, kind='bar')
```

```python
sns.catplot(x='class', y='fare', data=data, kind='box')
```
