# Grids
Grids are general types of plots that allow to map plot types to rows and columns of a grid.

```python
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

<hr>

## PairGrid
Pairgrid is a subplot grid for plotting pairwise relationship in a dataset.

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(sns.scatterplot)
```

```python
#Map to upper, lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

<hr>

## pairplot

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species', palette='rainbow')
```

<hr>

## Facet Grid
FacetGrid is the general way to create grids of plots based off of a feature.

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
```

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
g = g.map(sns.histplot, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time", row="smoker", hue="sex")
g = g.map(sns.scatterplot, "total_bill", "tip").add_legend()
```

<hr>

## JointGrid

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.histplot)
```
