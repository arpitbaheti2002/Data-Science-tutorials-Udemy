# Matplotlib Exercises
Follow the instructions to recreate the plots using the data:

### Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Import matplotlib.pyplot as plt**

```python
import matplotlib.pyplot as plt
```

## Exercise 1

**Follow alonf with these steps**
+ **Create a figure object called fig using plt.figure**
+ **Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.**
+ **Plot (x,y) on the axes and set the labels and titles to 'x', 'y' and 'title'.**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
ax.plot(x,y)
```

## Exercise 2

**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2.0.5,0.2,0.2] respectively**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

**Now plot(x,y) on both axes. And call your figure object to show it.**

```python
ax1.plot(x,y)
ax2.plot(x,y)

fig
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,0.4,0.4]**

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

**Now use x, y and z arrays to recreate the plot below. Notice the x limits and y limits on the inserted plot.**

```python
ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_xlim(0, 100)
ax1.set_ylim(0, 10000)

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)

fig
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
figure, axes = plt.subplots(nrows=1, ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
axes[0].plot(x,y,color="blue", ls="--", lw=3)
axes[1].plot(x,z,color="red", ls="-", lw=4)
figure
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots()**

```python
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,2))
axes[0].plot(x,y,color="blue", ls="-", lw=4)
axes[1].plot(x,z,color="red", ls="--", lw=3)
```
