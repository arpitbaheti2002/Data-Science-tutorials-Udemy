**Figure out how to get the stock data from Jan 1st 2006 to Jan 1st 2016 for each of these banks. Set each bank to be a seperate dataframe, with the variable name for that bank being its ticker symbol.**<br>
This will involve a few steps:

`# Bank of America`<br>
`BAC = data.get_data_yahoo('BAC', start=start, end=end)`


```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
```

```python
BAC = data.get_data_yahoo('BAC', start=start, end=end)
BAC.head()
```

```python
# CitiGroup (C)
C = data.get_data_yahoo('C', start=start, end=end)
C.head()
```

```python
# Goldman Sachs (GS)
GS = data.get_data_yahoo('GS', start=start, end=end)
GS.head()
```

```python
# JPMorgan Chase (JPM)
JPM = data.get_data_yahoo('JPM', start=start, end=end)
JPM.head()
```

```python
# Morgan Stanley (MS)
MS = data.get_data_yahoo('MS', start=start, end=end)
MS.head()
```

```python
# Wells Fargo (WFC)
WFC = data.get_data_yahoo('WFC', start=start, end=end)
C.head()
```

**Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

**Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list.**

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], keys=tickers, axis=1)
```

**Set the columns name levels**

```python
bank_stocks.columns.names = ['Bank Ticker', 'Stock Info']
```

```python
bank_stocks.head()
```

## EDA

**What is the the max Close price for each bank's stock throughout the time period?**

It is recommended to to check out the documentation on [Multi-Level Indexing](https://pandas.pydata.org/pandas-docs/stable/user_guide/advanced.html) and [Using .xs](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.xs.html)

```python
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()  
```

**Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock.**

```python
returns = pd.DataFrame()
```

**We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
returns = bank_stocks.xs(key='Close', level='Stock Info', axis=1).pct_change()
returns.head()
```

**Create a pairplot using seaborn of the returns dataframe.**

```python
sns.pairplot(returns)
```

**Using this returns DataFrame, figure out on what dates each bank stock had the best and worst single day returns. You should notice that 4 of the banks share the same day for the worst drop, did anything significant happen that day?**

```python
pd.DataFrame([returns.idxmin(), returns.idxmax()], index=['worst returns', 'best returns'])
```

**Take a look at the standard deviation of the returns, which stock would you classify as the riskiest over the entire time period?**

```python
returns.std()
# Citi Group Riskiest
```

**Which would you classify as the riskiest for the year 2015?**

```python
returns.loc['2015-01-01':'2015-12-31',:].std()
# Morgan Stanley and Bank of America
```

**Create a distplot using seaborn of the 2015 returns for Morgan Stanley**

```python
sns.displot(returns.loc['2015-01-01':'2015-12-31',:]['MS'], kde=True, bins=100, color='green')
```

**Create a distplot using seaborn of the 2008 returns for CitiGroup**

```python
sns.displot(returns.loc['2008-01-01':'2008-12-31',]['C'], bins=100, kde=True, color='red')
```

### Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

**Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
plt.figure(figsize=(12, 5), dpi=100)
sns.lineplot(bank_stocks['BAC']['Close'].loc['2008-01-01':'2008-12-31'], color='green', label='BAC close')
sns.lineplot(bank_stocks['BAC']['Close'].loc['2008-01-01':'2008-12-31'].rolling(30).mean(), label='30 Day Average')
```

```python
close_corr = bank_stocks.xs(key='Close', level='Stock Info', axis=1).corr()
sns.heatmap(close_corr, cmap='Reds', annot=True)
```

```python
close_corr.iplot(kind='heatmap', colorscale='Reds')
```

```python
sns.clustermap(close_corr, cmap='Reds', annot=True)
```
