**Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size.**

```python
df3.plot.scatter(x='a', y='b', figsize=(12,3), size=50, color='red')
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
```

**Create an area plot of all the columns for just the rows up to 30. (hint: use .ix)**

```python
df3.plot.area(alpha=0.5)
plt.xlim(0,30)
plt.ylim(0,3)
```

**Notice how the legend in our previous figure overlapped some of actual daigram. Can you figure out how to display the legend outside the plot.**

```python
df3.plot.area(alpha=0.5)
plt.xlim(0,30)
plt.ylim(0,3)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```
