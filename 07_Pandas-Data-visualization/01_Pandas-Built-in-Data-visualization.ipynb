## Style Sheets
+ Matplotlib [style sheets](https://matplotlib.org/2.0.2/gallery.html) can be used to make the plots look nicer. 
+ These style sheets include plot_bmh, plot_fivethirtyeight, plot_ggplot and more. Thses basically create a set of style rules that the plots follow.

**Before plt.style.use(), the plots look like this:**

```python
df1['A'].hist()
```

**After calling plt.style.use()** 

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
df1['A'].hist()
```

```python
plt.style.use('bmh')
df1['A'].hist()
```

```python
plt.style.use('dark_background')
df1['A'].hist()
```

```python
plt.style.use('fivethirtyeight')
df1['A'].hist()
```

```python
plt.style.use('ggplot')
```

## Plot Types
There are several plot types built-in to pandas, most of them are statistical by nature:
+ df.plot.area
+ df.plot.barh
+ df.plot.density
+ df.plot.hist
+ df.plot.line
+ df.plot.scatter
+ df.plot.bar
+ df.plot.box
+ df.plot.hexbin
+ df.plot.kde
+ df.plot.pie

Wecan also just call `df.plot(kind='hist')` and replace kind argument with any of the key terms.

<hr>

## Area

```python
df2.plot.area(alpha=0.4)
```

## Barplots

```python
df2.head()
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

## Histogram

```python
df1['A'].plot.hist(bins=50)
```

## Line Plots


```python
df1.plot.line(y='B', figsize=(12,3), lw=1)
```

## Scatter Plots

```python
df1.plot.scatter(x='A', y='B')
```

We can use c to color based off another column value.

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

Or we can use s to indicate size based off another column.

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*200)
```

## BoxPlots

```python
df2.plot.box()
# we can also pass by= argument for groupby
```

## Hexagonal Bin Plot

```python
df1.plot.hexbin(x='A', y='B', gridsize=25)
```

## Kernerl Density Estimation plot (KDE)

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```
